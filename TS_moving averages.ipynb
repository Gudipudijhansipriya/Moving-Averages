* Moving averge

```python
import numpy as np
  
window_size = 3
  
i = 0
moving_averages_3 = []
while i < len(df['Anomaly']) - window_size + 1:
    window_average = round(np.sum(df['Anomaly'][i:i+window_size]) / window_size, 2)
    moving_averages_3.append(window_average)
    i += 1
  
print("Moving average for three years",moving_averages_3)
```

```python
df['Anomaly'].rolling(window=3).mean().plot()
```
